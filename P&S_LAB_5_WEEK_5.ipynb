```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats




N = 1000
call_range = np.arange(0, 8)
probabilities = np.random.dirichlet(np.ones(len(call_range)), size=1)[0]

calls_per_day = np.random.choice(call_range, size=N, p=probabilities)

pmf_values = np.bincount(calls_per_day, minlength=len(call_range)) / N

cdf_values = np.cumsum(pmf_values)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(call_range, pmf_values, alpha=0.7, color='b', label='PMF')
plt.xlabel('Number of Calls per Day')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF)')
plt.xticks(call_range)
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(call_range, cdf_values, marker='o', linestyle='-', color='r', label='CDF')
plt.xlabel('Number of Calls per Day')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.xticks(call_range)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats




N_patients = 100
p_surgery = 0.25


prob_20 = stats.binom.pmf(20, N_patients, p_surgery)


prob_30_or_more = 1 - stats.binom.cdf(29, N_patients, p_surgery)


expected_patients = N_patients * p_surgery

print(f"Probability that exactly 20 patients need surgery: {prob_20:.4f}")
print(f"Probability that at least 30 patients need surgery: {prob_30_or_more:.4f}")
print(f"Expected number of patients requiring surgery per day: {expected_patients}")
```
