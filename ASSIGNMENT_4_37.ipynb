```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

file_path = '/content/car_data_250.csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
sns.histplot(df["Price"], bins=20, kde=True, color="blue")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Car Prices")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Horsepower", y="Mileage", hue="Fuel Type", palette="tab10", alpha=0.7)
plt.xlabel("Horsepower")
plt.ylabel("Mileage (MPG)")
plt.title("Horsepower vs. Mileage")
plt.legend(title="Fuel Type", bbox_to_anchor=(1, 1))
sns.regplot(data=df, x="Horsepower", y="Mileage", scatter=False, color="black", ci=None)
plt.show()

plt.figure(figsize=(12, 6))
brand_counts = df["Brand"].value_counts()
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette="viridis")
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.title("Number of Cars by Brand")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
fuel_year_counts = df.groupby(["Year", "Fuel Type"]).size().unstack()
fuel_year_counts.plot(kind="bar", stacked=False, figsize=(12, 6), colormap="tab10")
plt.xlabel("Manufacturing Year")
plt.ylabel("Number of Cars")
plt.title("Fuel Type Distribution Across Manufacturing Years")
plt.legend(title="Fuel Type")
plt.xticks(rotation=45)
plt.show()
```
